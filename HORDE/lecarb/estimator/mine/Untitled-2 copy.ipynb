```python
import pickle
def save_data_to_file(data,file_name):
    save_addr="./lecarb/estimator/mine/vec_data/"+file_name
    with open(save_addr, 'wb') as f:
        pickle.dump(data, f)
    print("the vec data has been stored in "+save_addr)

def load_data_from_pkl_file(file_name):
    load_addr="./lecarb/estimator/mine/vec_data/"+file_name
    with open(load_addr, 'rb') as f:
        data = pickle.load(f)
    print("has successfully loaded data from "+load_addr)
    return data
```

```python
import matplotlib.pyplot as plt
import numpy as np
load_addr="./distribution_data/new_50_"+"census13"+"_"+"original"+".pkl"
with open(load_addr, 'rb') as f:
    distribution_data = pickle.load(f)
print("has successfully loaded data from "+load_addr)

for attr_name,attr_distribution in distribution_data.items():
    # print(attr_name,attr_distribution)
    
    #
    #print(attr_distribution)
    
    [name,attr_type]=attr_name.split("\n")
    x=[]
    y=[]
    for i in attr_distribution:
        x.append(i[0])
        y.append(i[1])
    plt.plot(x, y, 'b*--', alpha=0.5, linewidth=1, label='census13')#'bo-'
    ## plot's','o''*'   ..., 

    plt.legend()  #label
    plt.xlabel(attr_name) #x_label
    plt.ylabel('frequency')#y_label
    
    #plt.ylim(-1,1)#y
    plt.show()
```
